```python
# Check number of cancelled and non-cancelled bookings
resort_hotel=df[df["hotel"]=="Resort Hotel"]
resort_hotel["is_canceled"].value_counts(normalize=True)
```

```python
City_hotel=df[df["hotel"]=="City Hotel"]
City_hotel["is_canceled"].value_counts(normalize =True)
```

```python
# Calculate average ADR by month for Resort Hotel
resort_hotel = (
    df[df["hotel"] == "Resort Hotel"]
    .groupby("arrival_date_month")[["adr"]]
    .mean()
)

# Calculate average ADR by month for City Hotel
city_hotel = (
    df[df["hotel"] == "City Hotel"]
    .groupby("arrival_date_month")[["adr"]]
    .mean()
)

# Arrange months in order
months = [
    "January","February","March","April",
    "May","June","July","August",
    "September","October","November","December"
]

resort_hotel = resort_hotel.reindex(months)
city_hotel = city_hotel.reindex(months)

# Plot graph
plt.figure(figsize=(15,6))

plt.plot(
    resort_hotel.index,
    resort_hotel["adr"],
    marker="o",
    label="Resort Hotel"
)

plt.plot(
    city_hotel.index,
    city_hotel["adr"],
    marker="o",
    label="City Hotel"
)

plt.title(
    "Average Daily Rate Comparison: City vs Resort Hotel"
)

plt.xlabel("Month")
plt.ylabel("Average Daily Rate (ADR)")

plt.xticks(rotation=45)
plt.legend()

plt.show()
```

```python
# Plot Average Daily Rate (ADR) comparison for cancelled and non-cancelled bookings

import matplotlib.pyplot as plt
import pandas as pd

# Convert date column
df["reservation_status_date"] = pd.to_datetime(
    df["reservation_status_date"]
)

# Cancelled bookings
cancelled_data = df[df["is_canceled"] == 1]

cancelled_df_adr = cancelled_data.groupby(
    "reservation_status_date"
)[["adr"]].mean()

cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values(
    "reservation_status_date",
    inplace=True
)

# Not cancelled bookings
not_cancelled_data = df[df["is_canceled"] == 0]

not_cancelled_df_adr = not_cancelled_data.groupby(
    "reservation_status_date"
)[["adr"]].mean()

not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values(
    "reservation_status_date",
    inplace=True
)

# Plot
plt.figure(figsize=(20,6))

plt.plot(
    not_cancelled_df_adr["reservation_status_date"],
    not_cancelled_df_adr["adr"],
    label="Not Cancelled"
)

plt.plot(
    cancelled_df_adr["reservation_status_date"],
    cancelled_df_adr["adr"],
    label="Cancelled"
)

plt.title("Average Daily Rate")
plt.xlabel("Reservation Status Date")
plt.ylabel("ADR")

plt.xticks(rotation=45)
plt.legend()
plt.show()
```
